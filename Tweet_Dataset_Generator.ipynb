```
def extract_timeline_as_df(timeline_list):

    columns = set()
    allowed_types = [str, int]
    tweets_data = []

    for status in timeline_list:

        status_dict = dict(vars(status))
        keys = status_dict.keys()
        single_tweet_data = {"user": status.user.screen_name, "author": status.author.screen_name}

        for k in keys:
            try:
                v_type = type(status_dict[k])
            except:
                v_type = None

            if v_type != None:
                if v_type in allowed_types:
                    single_tweet_data[k] = status_dict[k]
                    columns.add(k)

        tweets_data.append(single_tweet_data)


    header_cols = list(columns)
    header_cols.append("user")
    header_cols.append('author')
    df = pd.DataFrame(tweets_data, columns=header_cols)
    return df

```

```
def extract_timeline_as_df2(timeline_list):

    columns = set()
    allowed_types = [str, int]
    tweets_data = []

    for status in timeline_list:

        status_dict = dict(vars(status))
        keys = status_dict.keys()
        single_tweet_data = {"user": status.user.screen_name, "author": status.author.screen_name}

        for k in keys:
            try:
                v_type = type(status_dict[k])
            except:
                v_type = None

            if v_type != None:
                if v_type in allowed_types:
                    single_tweet_data[k] = status_dict[k]
                    columns.add(k)

        tweets_data.append(single_tweet_data)


    header_cols = list(columns)
    header_cols.append("user")
    header_cols.append('author')
    df = pd.DataFrame(tweets_data, columns=header_cols)
    return df



def user_tweets(user_names, tweet_per_user = 20):

    df_ = pd.DataFrame()
    # required pages for tweet_count (20 tweets per page)
    pages = int(np.ceil(tweet_per_user/20))
    
    for user_name in user_names:
        user_vars_ = api.get_user(user_name)

        # create df of each page and append
        for i in range(1, pages+1):

            temp_user_timeline = user_vars_.timeline(page = i)
            page_df = extract_timeline_as_df2(temp_user_timeline)

            df_ = df_.append(page_df, ignore_index = True)

    return df_
```
